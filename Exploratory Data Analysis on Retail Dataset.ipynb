```python
# Visualize data
sns.pairplot(ssd)
```

```python
# Heatmap for visualize correlations 
fig,axes = plt.subplots(1,1,figsize=(12,7))
sns.heatmap(ssd.corr())
plt.show()
```

```python
# Profit vs. Sales visualize
fig, axes = plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total Profit vs Sales")
sns.barplot(data=ssd.groupby('Sub-Category')['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axes[1])
ssd.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
# Sales vs Quantity visualize
fig,axes = plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total Sales vs Quantity")
sns.barplot(data=ssd.groupby('Sub-Category')['Sales','Quantity'].agg(sum),x='Sales',y='Quantity',ax=axes[1])
ssd.groupby('Sub-Category')['Sales','Quantity'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
# Discount vs Quantity visualize
fig,axes = plt.subplots(1,2,figsize=(14,5))
ssd.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title("Quantity & Profit Relation based on sub-category")
ssd.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[1]).set_title("Discount & Profit Relation based on sub-category")
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution  Data
fig,axes = plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Distribution plots", fontsize=16)
sns.distplot(ssd['Sales'],ax=axes[0,0])
sns.distplot(ssd['Profit'],ax=axes[0,1])
sns.distplot(ssd['Discount'],ax=axes[1,0])
sns.distplot(ssd['Quantity'],ax=axes[1,1])
plt.show()
```

```python
# Quantity based on Region
fig,axes = plt.subplots(1,1,figsize=(12,7))
sns.countplot(ssd['Quantity'],hue=ssd['Region'])
plt.show()
```

```python
# Sales with different shipping modes and segments
fig,axes = plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Sales with different shipping modes and segments", fontsize=16)
sns.barplot(ssd['Ship Mode'],ssd['Sales'],ax=axes[0,0])
sns.lineplot(ssd['Ship Mode'],ssd['Sales'],ax=axes[0,1])
sns.barplot(ssd['Segment'],ssd['Sales'],ax=axes[1,0])
sns.lineplot(ssd['Segment'],ssd['Sales'],ax=axes[1,1])
plt.show()
```

## Some important findings
- The features Profit and Discounts are highly related.
- Over less quantity of products also the sales were high.
- The maximum quantity of product in demand was in range.
- The mode of shipping doesn't affect much to the sales.
- The Home Office provides highest sales followed by corporate by a sllght variation.
